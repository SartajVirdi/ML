#PCA (Pricipal Component Analysis)

---

Applications:

1) Visualisation (Reduction of number of features)

2) Data Compression

3) Speeding up training of supervised learning model.

4) Reduction of storage and transmission costs.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([[1, 1], [2, 1], [3, 2], [-1, -1], [-2, -1], [-3, -2]]) #synthetic data
pca_1 = PCA(n_components=1)
pca_1.fit(X) #fit() incldues mean normalisation + prinicpal axis
pca_1.explained_variance_ratio_
X_trans_1 = pca_1.transform(X)
X_reduced_1 = pca_1.inverse_transform(X_trans_1)
print('X(x1, x2) transformeed into z :')
print(X_trans_1)
print('Reverse transformation back to original dataset with offsets noticeable: ')
print(X_reduced_1)
print('Original for comparison: ')
print(X)
```

```python
#original plot using 2 features
x = X[:, 0]
y = X[:, 1]
plt.scatter(x, y, c='Blue', marker = 'x')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Dataset with 2 features')
```

```python
#plot with principal axis (z)
l = X_reduced_1[:, 0]
m = X_reduced_1[:, 1]
plt.scatter(l, m, color = 'red', marker = 'o')
plt.title('Inverse transformed (Z axis scatter points)')
k = np.array([[0, 0]])
plt.scatter(k[:, 0], k[:, 1], color = 'black', marker = 'o', s = 100)
print(l)
print(m)
plt.plot([l[2], l[5]], [m[2], m[5]], color = 'green', linestyle = '--')
```

```python
X = np.array([[1, 1], [2, 1], [3, 2], [-1, -1], [-2, -1], [-3, -2]]) #synthetic data
pca_1 = PCA(n_components=2) #when number of components set to 2, the reverse transform gives us original dataset
pca_1.fit(X) #fit() incldues mean normalisation + prinicpal axis
pca_1.explained_variance_ratio_
X_trans_1 = pca_1.transform(X)
X_reduced_1 = pca_1.inverse_transform(X_trans_1)
print('X(x1, x2) transformeed into z1, z2 :')
print(X_trans_1)
print('Reverse transformation back to original dataset with offsets noticeable: ')
print(X_reduced_1)
print('Original for comparison: ')
print(X)
```
